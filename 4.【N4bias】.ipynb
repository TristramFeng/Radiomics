```python
import SimpleITK as sitk
import numpy as np
from my_MRI_tool_kit import DirectoryReader_dict, convert_nifti_datatype
from tqdm import tqdm
import os
```

```python
# ################################################
# N4偏置场矫正(指定mask文件)
# ################################################
def N4BiasFieldCorrection_mask(image_path, mask_path, output_path):
    # 读nifty
    input_image = sitk.ReadImage(image_path)
    
    # 设置蒙版，只对蒙版内的部分操作
    mask_image = sitk.ReadImage(mask_path)
    input_image = sitk.Cast(input_image, sitk.sitkFloat32)
    
    # 主运行
    corrector = sitk.N4BiasFieldCorrectionImageFilter() # N4偏置场
    output_image = corrector.Execute(input_image,mask_image)
    output_image = sitk.Cast(output_image, sitk.sitkInt16)    # sitk.sitkInt16
    
    sitk.WriteImage(output_image, output_path)
        
    print(f'【N4 processed successfully】{image_path}')

# N4BiasFieldCorrection_mask('G:\________b_regd_rsmp_T1CE.nii.gz',
#                            'G:\____brain_mask.nii.gz',
#                            'G:\111.nii.gz',)
```

```python
for key,value in tqdm(directories.items()):
    print(key)

    T1 = '________b_regd_rsmp_T1.nii.gz'
    T1CE = '________b_regd_rsmp_T1CE.nii.gz'
    T2 = '________b_regd_rsmp_T2.nii.gz'
    T2Flair = '________b_regd_rsmp_T2Flair.nii.gz'
    fDWI_2 = '________b_regd_rsmp_fDWI_2.nii.gz'
    ADC = '________b_regd_rsmp_ADC.nii.gz'
    ADC_manual = '________b_regd_rsmp_ADC_manual.nii.gz'
    
    for i in [T1, T1CE, T2, T2Flair, fDWI_2, ADC, ADC_manual]:
        image_path = os.path.join(key,i)
        mask_path = os.path.join(key,'____brain_mask.nii.gz')
        output_path = os.path.join(key,'________n4_'+i[8:])
        
        if os.path.exists(image_path):
            N4BiasFieldCorrection_mask(image_path,mask_path,output_path)
    
```
